### Clean up of downloaded data from GeyserTimes.com and geyserstudy.org

```python
import pandas as pd
import datetime as dt
import re
```

```python
df = pd.read_csv('old_faithful.csv', index_col='eruptionID')
df.head()
```

```python
df.columns
```

```python
df.drop(['has_seconds', 'exact', 'ns', 'ie', 'E', 'A', 'wc', 'entrant','ini', 'maj', 'min', 'q','eruption_comment', 'time_updated', 'time_entered',
       'associated_primaryID', 'other_comments','Old_Faithful_Preplay_Time_VEC','Old_Faithful_Height_VEC'], axis=1, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['eruption_time_epoch'], unit='s')
```

```python
df.head()
```

```python
df['year'] = df['date'].dt.year
df.head()
```

```python
df["waiting_time_mins"] = df["eruption_time_epoch"].diff(-1)/60
df.tail()
```

```python
df["waiting_time_hrs"] = df["eruption_time_epoch"].diff(-1)/3600
df.head()
```

```python
df['dur'] = df['duration'].str.findall('\d+')
```

```python
df.head()
```

```python
df.to_csv('geyser_times_data.csv', encoding='utf-8', index=True)
```

```python
combined_df = pd.read_csv('real_master.csv', index_col='eruptionID')
```

```python
combined_df.info()
```

```python
combined_df.head()
```

```python
df_waiting = combined_df.copy()
```

```python
df_waiting['waiting_min']=pd.to_numeric(df_waiting['waiting_min'],errors='coerce')
df_waiting.info()
```

```python
df_waiting = df_waiting.loc[~df_waiting['waiting_min'].isna()]
df_waiting.info()
```

```python
df_wait_filt = df_waiting[(df_waiting['waiting_min'] < 200)]
```

```python
filt_df = df_wait_filt[(df_wait_filt['waiting_min']> 2)]
```

```python
filt_df.count()
```

```python
filt_df.to_csv('waiting_filtered.csv', encoding='utf-8', index=True)
```

```python
duration_df = filt_df.dropna()
```

```python
duration_df.count()
```

```python
duration_df.to_csv('duration.csv', encoding='utf-8', index=True)
```
